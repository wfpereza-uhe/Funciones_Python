```python
def interes_simple(capital=0, tasa=0.05, tiempo_meses=12):
  """
  calcula el interes simple
  parametros:
    capital(float): monto de dinero
    tiempo_meses(int): tiempo en meses
    tasa(float): tasa de interes
  retorna:
    interes(float): interes simple
  """
  interes = capital*tasa*(tiempo_meses/12)
  return interes
```

```python
def dividir(numero1,numero2):
  try:
    resultado = numero1/numero2
  except ZeroDivisionError:
    print("No se puede dividir por cero")
    resultado = 0
  except:
    print("No se puede dividir")
    resultado = 0
  return resultado
```
